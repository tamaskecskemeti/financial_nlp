```python
from transformers import AutoModelForTokenClassification, AutoTokenizer, pipeline, BertLMHeadModel
import torch
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
device
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model = AutoModelForTokenClassification.from_pretrained('dslim/bert-large-NER').to(device)
tokenizer = AutoTokenizer.from_pretrained('dslim/bert-large-NER')
pipe = pipeline(
    "ner",
    model=model,
    tokenizer=tokenizer,
    device_map="auto",
)
```

```python
text = """Return a list of named entities in the text.
Text: The Golden State Warriors are an American professional basketball team based in San Francisco.
Named entities:
"""
```

```python
sequences = pipe(text)
```

```python
for seq in sequences:
  print(f"{seq['word']}")
```
